```python
def load_session_betas(dataset, subj, sess, version, zscore = True):
    
    print(f'loading {dataset} betas: {subj}, sess {sess}, {version}, zscore = {zscore}')
    if dataset == 'B5K':
        betadir = '/lab_data/tarrlab/common/datasets/BOLD5000_GLMsingle/betas/BOLD5000_complete'
        
        betafn = f"{betadir}/{subj}_GLMbetas-{version}_ses-{sess:02d}.nii.gz"
        
        # load betas
        print(f'\t{betafn}')
        assert(exists(betafn))
        betas = nib.load(betafn).get_data()
    
    elif dataset == 'NSD':
        betadir = '/lab_data/tarrlab/jacobpri/BOLD5000-GLMs/betas/09_08_20_one-sess_NSD'
        
        if 'ASSUMEHRF' in version:
            betadir = f'{betadir}_assume'
            version = 'TYPEB-FITHRF'
            
        betafn = f"{betadir}/{subj}/session{sess:02d}/{version.replace('-','_')}.mat"
        
        # load betas
        print(f'\t{betafn}')
        assert(exists(betafn))
        betas = pymatreader.read_mat(betafn)['modelmd']
    
    # zscore if necessary
    if zscore:
        betas = stats.zscore(betas,axis=3)
            
    print(f'\tbetas successfully loaded. shape: {betas.shape}')
    
    return betas
    
#betas = load_session_betas('B5K','CSI1', 15, 'TYPEC-FITHRF-GLMDENOISE')
#betas = load_session_betas('NSD','subj01', 10, 'TYPEA-ASSUMEHRF')
```

```python
def compute_autocorr_fn(rdm, maxlag = 50):
    assert(rdm.shape[0] == rdm.shape[1])
    assert(np.sum(np.isnan(rdm)) == 0)
    
    summary = []
    for lag in range(0,maxlag+1):
        corrs = []
        for i in range(rdm.shape[0]-lag):
            corrs.append(rdm[i+lag,i])
        summary.append(np.mean(corrs))
    return summary
            
```

```python
# temporal autocorrelation analysis 

rels = [0,0.3]

for ds in tqdm(['NSD','B5K']):
    
    if ds == 'B5K':
        ds_ = 'BOLD5000'
        nses = 15
    else:
        ds_ = 'NSD'
        nses = 10
        
    if ds == 'NSD':
        subjs = [f'subj0{n}' for n in range(1,5)]
    else:
        subjs = [f'CSI{n}' for n in range(1,5)]
    
    for subj in tqdm(subjs):
        
        if subj == 'CSI4':
            nses = 9
        
        print(ds,subj)
        
        # get version-averaged reliability volume to use for ranking voxels
        reliability_vol = reliability[ds][subj]['mean']
         
        for v in tqdm(range(len(beta_versions))):
            
            beta_version = beta_versions[v]
            rel_version = rel_versions[v]
            
            # iterate through versions, sessions
            for ses in range(1,nses + 1):
                            
                if ses < 10:
                    sesstr = '0' + str(ses)
                else:
                    sesstr = str(ses)
                
                
                #print('checking to see if betas need to be loaded')
                
                # check to see if betas need to be loaded
                loadbetas = False
                for rel in rels:
                    savedir = join(homedir, 'sessionRDMs', ds_, subj, rel_version)
                    
                    if not exists(savedir):
                        os.makedirs(savedir,exist_ok=True)
                    
                    # if at least one file needs to be computed, load the session betas
                    savefn = join(savedir, 'RDM_session' + sesstr + '_relthresh_' + str(rel).replace('.','_') + '.npy')
                    if not exists(savefn):
                        loadbetas = True

                if loadbetas is True:
                    
                    print('loading betas')
        
                    betas = load_session_betas(ds, subj, ses, beta_version.replace('_','-'))
            
                    mask = masks[ds][subj]
                
                    betas = betas[mask == 1]
                    print(betas.shape)

                    # threshold at each reliability value
                    for rel in rels:
                        
                        print(ds,subj,rel_version,sesstr,rel)

                        savefn = join(savedir, 'RDM_session' + sesstr + '_relthresh_' + str(rel).replace('.','_') + '.npy')
                        
                        if not exists(savefn):
                            
                            # compute RDM
                            #mask = np.logical_and(np.logical_not(np.isnan(betas[:,:,:,0])), (reliability_vol > rel))
                            #set_trace()
                            #mask = np.logical_and(masks[ds][subj] == 1, reliability_vol > rel)
                            rel_mask = reliability_vol > rel
                            betas_2D = betas[rel_mask].T
                            print(betas_2D.shape)
                            
                            if np.sum(np.isnan(betas_2D)) > 0:
                                badvox = np.unique(np.argwhere(np.isnan(betas_2D))[:,1])
                                goodvox = np.setdiff1d(np.arange(betas_2D.shape[1]), badvox)
                                betas_2D = betas_2D[:,goodvox]

                            assert(np.sum(np.isnan(betas_2D)) == 0)

                            RDM = squareform(pdist(betas_2D,'correlation'))
                            print('saving RDM\n', savefn)
                            np.save(savefn, RDM)

                else:
                    #print('all files already computed. done.\n')
                    assert(1==1)
                    
                
                
```
